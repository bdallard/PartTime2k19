On a créé un tableau 2D contenant des valeurs aléatoires.

```python
#np.sum(rdm_mat)
rdm_mat.sum()
```

Faire en sorte de récupérer la valeur maximale sur chaque ligne (sans utiliser ni fonction ni boucle) 

```python
np.max(rdm_mat, axis=1)
```

Faire la même chose en définissant un tableau avec une boucle for et la fonction max de numpy

```python
[np.max(rdm_mat[i]) for i in range(0,len(rdm_mat))]
#avec une fonction
def maxCol(array): 
    tab = []
    for i in range(0,len(array)):
        tab.append(np.max(array[i]))
    return tab 
```

Faire de même pour récupérer la valeur maximum pour chaque colonne (sans boucles)

```python
np.max(rdm_mat, axis=0)
```
